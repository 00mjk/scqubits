```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import scqubits as qubit

import numpy as np
import matplotlib.pyplot as plt
```

```python
rhombus = qubit.Rhombus(
    EJ1 = 20.0,
    EJ2 = 20.0,
    EC1 = 0.5,
    EC2 = 0.5,
    ECS = 0.5,
    ng1 = 0.0,
    ng2 = 0.0,
    ngs = 0.0,
    flux = 0.5,
    kbt = 3,
    N1 = 10,
    N2 = 10,
    NS = 10
)
```

```python
eng = rhombus._evals_calc(2)
eng[1] - eng[0]
```

```python
esys = rhombus.eigensys()
fig, axes = rhombus.plot_phi_1_phi_2_wavefunction(esys, which=0, mode='abs', zero_calibrate=True, figsize=(4,4))
axes.set_ylabel(r'$\phi_1$')
axes.set_xlabel(r'$\phi_2$')
fig, axes = rhombus.plot_phi_1_theta_wavefunction(esys, which=0, mode='abs', zero_calibrate=True, figsize=(4,4))
axes.set_ylabel(r'$\phi_1$')
axes.set_xlabel(r'$\phi_2$')
fig, axes = rhombus.plot_phi_2_theta_wavefunction(esys, which=0, mode='abs', zero_calibrate=True, figsize=(4,4))
axes.set_ylabel(r'$\phi_1$')
axes.set_xlabel(r'$\phi_2$')
```

```python
rhombus.plot_potential()
```

```python
ngs_list = np.linspace(0, .5, 20)
rhombus.plot_evals_vs_paramvals('ngs', ngs_list, evals_count=2, subtract_ground=True, figsize=(4,4));
```

```python
ng1_list = np.linspace(0, .5, 50)
rhombus.plot_evals_vs_paramvals('ng1', ng1_list, evals_count=2, subtract_ground=True, figsize=(4,4));
```

```python
esys = rhombus.eigensys()
ft_wfnc0 = rhombus.ft_wavefunction(esys, which=0)
ft_wfnc1 = rhombus.ft_wavefunction(esys, which=1)
```

```python
fig, axes = rhombus.plot_n_theta_n_phi_2_wavefunction(ft_wfnc0, mode='abs', zero_calibrate=True, figsize=(4,4))
axes.set_ylabel(r'$n_\theta$')
axes.set_xlabel(r'$n_{\phi 2}$')
axes.set_yticks(np.arange(-10,11,2))
axes.set_xticks(np.arange(-10,11,2))
fig, axes = rhombus.plot_n_theta_n_phi_2_wavefunction(ft_wfnc1, mode='abs', zero_calibrate=True, figsize=(4,4))
axes.set_ylabel(r'$n_\theta$')
axes.set_xlabel(r'$n_{\phi 2}$')
axes.set_yticks(np.arange(-10,11,2))
axes.set_xticks(np.arange(-10,11,2))
```

```python
fig, axes = rhombus.plot_n_phi_1_n_theta_wavefunction(ft_wfnc0, mode='abs', zero_calibrate=True,figsize=(4,4))
axes.set_ylabel(r'$n_{\phi 1}}$')
axes.set_xlabel(r'$n_{\theta}$')
axes.set_yticks(np.arange(-10,11,2))
axes.set_xticks(np.arange(-10,11,2))
fig, axes = rhombus.plot_n_phi_1_n_theta_wavefunction(ft_wfnc1, mode='abs', zero_calibrate=True,figsize=(4,4))
axes.set_ylabel(r'$n_{\phi 1}}$')
axes.set_xlabel(r'$n_{\theta}$')
axes.set_yticks(np.arange(-10,11,2))
axes.set_xticks(np.arange(-10,11,2))
```

```python
fig, axes = rhombus.plot_n_phi_1_n_phi_2_wavefunction(ft_wfnc0, mode='abs', zero_calibrate=True,figsize=(4,4))
axes.set_ylabel(r'$n_{\phi 1}}$')
axes.set_xlabel(r'$n_{\phi 2}$')
axes.set_yticks(np.arange(-10,11,2))
axes.set_xticks(np.arange(-10,11,2))
fig, axes = rhombus.plot_n_phi_1_n_phi_2_wavefunction(ft_wfnc1, mode='abs', zero_calibrate=True,figsize=(4,4))
axes.set_ylabel(r'$n_{\phi 1}}$')
axes.set_xlabel(r'$n_{\phi 2}$')
axes.set_yticks(np.arange(-10,11,2))
axes.set_xticks(np.arange(-10,11,2))
```
