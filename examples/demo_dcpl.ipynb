```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import scqubits as qubit

import numpy as np
```

# Double Cooper pair tunneling Qubit with addtional inductor

$H_\text{dcp}=4E_\text{C}[2n_\phi^2+\frac{1}{2}(n_\varphi-N_\text{g}-n_\theta)^2+xn_\theta^2]+E_\text{L}(\frac{1}{4}\phi^2+\theta^2)-2E_\text{J}\cos(\varphi)\cos(\frac{\phi}{2}+\frac{\varphi_\text{ext}}{2})+E_\text{LA}(\frac{\varphi_\text{ext}}{2}-\varphi-\theta)^2$

```python
dcpl_qubit = qubit.Dcpl(
    EJ = 15.0,
    EC = 2.0,
    EL = 1.0,
    ELA = 1.0,
    x = 0.02,
    Ng = 0.0,
    flux = 0.5,
    N0 = 60,
    q0 = 40,
    p0 = 10
)
```

```python
print(dcpl_qubit)
```

```python
spectrum = dcpl_qubit.get_spectrum_vs_paramvals('flux',[0],evals_count=14,subtract_ground=True)
spectrum.energy_table
```

```python
esys = dcpl_qubit.eigensys()
fig, axes = dcpl_qubit.plot_n_phi_n_varphi_wavefunction(esys,mode='abs',which=1,figsize=(4,4))
axes.set_ylabel(r'$N_\phi$')
axes.set_xlabel(r'$N_\theta$')
axes.set_yticks([-2,-1.5, -1,-0.5, 0,0.5, 1,1.5, 2])
axes.set_yticklabels(['-4','-3','-2','-1','0','1','2','3','4'])
axes.set_xticks([-4,-3, -2,-1, 0,1,2,3,4])
axes.set_xlim([-4,4])
axes.set_ylim([-2,2])

```

```python
flux_list = np.linspace(-.1, .6, 50)
dcpl_qubit.plot_evals_vs_paramvals('flux', flux_list, evals_count=14, subtract_ground=True, figsize=(4,4));
```
